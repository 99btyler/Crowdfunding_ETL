# Part 1: Extract and Transform

```python
import pandas as pd
pd.set_option('max_colwidth', 400)

import numpy as np
```

## Extract data from crowdfunding.xlsx

```python
crowdfunding_df = pd.read_excel("resources/input/crowdfunding.xlsx")
crowdfunding_df.head()
```

```python
crowdfunding_df.info()
```

### Create the DataFrames and CSV files for "Category" and "Subcategory"

```python
crowdfunding_df.columns
```

```python
crowdfunding_df[["category", "subcategory"]] = crowdfunding_df["category & sub-category"].str.split("/", expand=True)

crowdfunding_df.head()
```

```python
categories = crowdfunding_df["category"].unique()
subcategories = crowdfunding_df["subcategory"].unique()

print(f"categories: {categories}")
print(f"subcategories: {subcategories}")
```

```python
num_of_categories = len(categories)
num_of_subcategories = len(subcategories)

print(f"num_of_categories: {num_of_categories}")
print(f"num_of_subcategories: {num_of_subcategories}")
```

```python
category_id_array = np.arange(1,num_of_categories+1)
subcategory_id_array = np.arange(1,num_of_subcategories+1)

print(f"category_id_array: {category_id_array}")
print(f"subcategory_id_array: {subcategory_id_array}")
```

```python
category_ids = [f"category{i}" for i in range(1,num_of_categories+1)]
subcategory_ids = [f"subcategory{i}" for i in range(1,num_of_subcategories+1)]

print(f"category_ids: {category_ids}")
print(f"subcategory_ids: {subcategory_ids}")
```

```python
category_df = pd.DataFrame({
    "category_id": category_ids,
    "category": categories
})

category_df
```

```python
subcategory_df = pd.DataFrame({
    "subcategory_id": subcategory_ids,
    "subcategory": subcategories
})

subcategory_df
```

```python
category_df.to_csv("resources/output/category.csv", index=False)
subcategory_df.to_csv("resources/output/subcategory.csv", index=False)
```

### Create the DataFrames and CSV files for "Campaign"

```python
campaign_df = crowdfunding_df.copy()
campaign_df.head()
```

```python
campaign_df = campaign_df.rename(columns={
    "blurb": "description",
    "launched_at": "launched_date",
    "deadline": "end_date"
})

campaign_df.head()
```

```python
campaign_df["goal"] = campaign_df["goal"].astype(float)
campaign_df["pledged"] = campaign_df["pledged"].astype(float)

campaign_df.head()
```

```python
campaign_df.info()
```

```python
campaign_df["launched_date"] = pd.to_datetime(campaign_df["launched_date"], unit="s").dt.strftime("%Y-%m-%d")
campaign_df["end_date"] = pd.to_datetime(campaign_df["end_date"], unit="s").dt.strftime("%Y-%m-%d")

campaign_df.head()
```

```python
campaign_df.head()
```

```python
merged_campaign_df = pd.merge(campaign_df, category_df, on="category")
merged_campaign_df = pd.merge(merged_campaign_df, subcategory_df, on="subcategory")

merged_campaign_df.tail(10)
```

```python
merged_campaign_df.drop(["staff_pick", "spotlight", "category & sub-category", "category", "subcategory"], axis=1, inplace=True)
merged_campaign_df.head()
```

```python
merged_campaign_df.to_csv("resources/output/campaign.csv", index=False)
```
